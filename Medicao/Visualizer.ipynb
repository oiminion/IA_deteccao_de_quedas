```python
import pandas as pd

VLibrasil_df = pd.read_csv("E:/Unesp/TCC/Codigo/Metricas/data_from_video_VLibrasil.csv")
MINDS_df = pd.read_csv("E:/Unesp/TCC/Codigo/Metricas/data_from_video_MINDS.csv")

df = pd.concat([VLibrasil_df, MINDS_df], ignore_index=True)

df.head()
```

```python
import matplotlib.pyplot as plt

#visualização da quantidade de vídeos com detecção menor que um certo valor.

cols = [
    'percent_detected_left_hand_between_first_and_last',
    'percent_detected_right_hand_between_first_and_last',
    'percent_detected_both_hands_between_first_and_last'
]

quantity = []
detection = []
for i in range(20):
    aux = df[(df[cols].max(axis=1) < ((i+1)*0.05)) & (df[cols].max(axis=1) >= (i*0.05))]
    quantity.append(len(aux))
    detection.append(i*0.05)

plt.bar(detection, quantity, width=0.04)

# Add descriptive labels
plt.xlabel('Detection')
plt.ylabel('Quantity')
```

```python
import matplotlib.pyplot as plt
result = []
for index, row in df.iterrows():
    min_value = min(row['first_frame_with_left_hand'], row['first_frame_with_right_hand'], row['first_frame_with_both_hands'])
    max_value = max(row['last_frame_with_left_hand'], row['last_frame_with_right_hand'], row['last_frame_with_both_hands'])
    if min_value == 0 and max_value == 0:
        result.append(0)
    else:
        result.append(max_value - min_value)

result = pd.DataFrame(result, columns=['duration_frames'])
result.describe()

print(result['duration_frames'].quantile(0.96))

result = result.sort_values(by='duration_frames', ascending=False)
result['duration_frames'].plot(kind='bar',figsize=(20, 5), width=1)
plt.show()
```
